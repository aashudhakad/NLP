```python
def edit_distance(word1, word2):
    n = len(word1)
    m = len(word2)

    # DP table
    dp = [[0] * (m + 1) for _ in range(n + 1)]

    # Base cases
    for i in range(n + 1):
        dp[i][0] = i   # deletions

    for j in range(m + 1):
        dp[0][j] = j   # insertions

    # Fill DP table
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(
                    dp[i - 1][j]+1,     # deletion
                    dp[i][j - 1]+1,     # insertion
                    dp[i - 1][j - 1]+2  # substitution
                )

    return dp[n][m]


# Input
word1 = input("Enter first word: ")
word2 = input("Enter second word: ")

# Output
result = edit_distance(word1, word2)
print("Minimum Edit Distance =", result)
```
