```python
import torch
import numpy as np

tensor_from_list = torch.tensor([1, 2, 3])
print(tensor_from_list)

np_array = np.array([[1, 2], [3, 4]])
tensor_from_numpy = torch.from_numpy(np_array)
print(tensor_from_numpy)
```

```python
import torch

a = torch.tensor([1, 2, 3])
b = torch.tensor([1, 2, 3])

addition = a + b
print(addition)

multiplication = a * b
print(multiplication)
```

```python
import torch
import numpy as np

t = torch.tensor([1, 2, 3])
print(t.shape)
print(t.dtype)

A = torch.tensor([[1, 2],
                  [3, 4]])
B = torch.tensor([[5, 6],
                  [7, 8]])
print(torch.mm(A, B))

x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])
print(torch.dot(x, y))

T = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])
print(T[0])
print(T[1])
print(T[0][1])
print(T[1][2])

print(torch.empty(2, 3))
print(torch.zeros(2, 3))
print(torch.rand(2, 3))
print(torch.ones(2, 3))
print(torch.eye(3))
print(torch.arange(0, 10, 2))
print(torch.linspace(0, 1, 5))
print(torch.add(torch.tensor([1, 2, 3]), torch.tensor([4, 5, 6])))
print(torch.mm(torch.tensor([[1, 2]]), torch.tensor([[3], [4]])))
print(torch.dot(torch.tensor([1, 2]), torch.tensor([3, 4])))
print(torch.bmm(torch.rand(2, 3, 4), torch.rand(2, 4, 5)))
print(torch.abs(torch.tensor([-1, -2, 3])))
print(torch.argmax(torch.tensor([10, 20, 5])))
print(torch.argmin(torch.tensor([10, 20, 5])))
print(torch.mean(torch.tensor([1.0, 2.0, 3.0])))
print(torch.sort(torch.tensor([3, 1, 2])))
print(torch.cat((torch.tensor([1, 2]), torch.tensor([3, 4]))))
```
